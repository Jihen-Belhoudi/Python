```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")


class Member(Person):
    def __init__(self, name, age, member_id):
        super().__init__(name, age)
        self.member_id = member_id
        self.books_borrowed = []

    def borrow_book(self, book):
        self.books_borrowed.append(book)
        print(f"{self.name} borrowed '{book}'.")

    def return_book(self, book):
       if book in self.books_borrowed:
            self.books_borrowed.remove(book)
            print(f"{self.name} returned '{book}'.")
       else:
            print(f"{self.name} does not have '{book}' borrowed.")

    def display_books_borrowed(self):
        print(f"{self.name}'s Books Borrowed:")
        for book in self.books_borrowed:
            print(f" {book}")

class Library:
    def __init__(self):
        self.members = []
        self.librarians = []
        self.books_collection = []

    def add_member(self, member):
        self.members.append(member)
        print(f"Member '{member.name}' added to the library.")

    def add_librarian(self, librarian):
        self.librarians.append(librarian)
        print(f"Librarian '{librarian.name}' added to the library.")

    def display_library_info(self):
        for member in self.members:
            member.display_info()
        for librarian in self.librarians:
            librarian.display_info()
        for book in self.books_collection:
            print(book)


class Librarian(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

    def add_book(self, book, library):
        library.books_collection.append(book)
        print(f"Librarian {self.name} added '{book}' to the collection.")

    def remove_book(self, book, library):
        if book in library.books_collection:
            library.books_collection.remove(book)
            print(f"Librarian {self.name} removed '{book}' from the collection.")
        else:
            print(f"'{book}' not found in the collection.")

    def display_library_info(self, library):
        print("\nLibrary Information:")
        print("Members:")
        for member in library.members:
            member.display_info()
        print("\nLibrarians:")
        for librarian in library.librarians:
            librarian.display_info()
        print("\nBooks in Collection:")
        for book in library.books_collection:
            print(book)



# Example Usage:
library = Library()

# Create Members
member1 = Member("Jihen", 25, "M001")
member2 = Member("Sabrine", 30, "M002")
member3 = Member("Omar", 30, "M003")
# Create Librarians
librarian1 = Librarian("Jacem", 40, "L001")
librarian2 = Librarian("Amine", 50, "L002")
# Add Members and Librarians to the Library
library.add_member(member1)
library.add_member(member2)
library.add_librarian(librarian1)

# Librarian Adds Books to the Library
librarian1.add_book("Book1", library)
librarian1.add_book("Book2", library)

# Members Borrow and Return Books
member1.borrow_book("Book1")
member2.borrow_book("Book2")
member1.display_books_borrowed()
member2.display_books_borrowed()
member1.return_book("Book1")
member2.return_book("Book3")

# Display Library Information
library.display_library_info()
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")


class Student(Person):
    def __init__(self, name, age, student_id, grade):
        super().__init__(name, age)
        self.student_id = student_id
        self.grade = grade

    def calculate_average(self):
        print(f"Student {self.name}'s average grade: {self.grade}")


class Teacher(Person):
    def __init__(self, name, age, employee_id, subject):
        super().__init__(name, age)
        self.employee_id = employee_id
        self.subject = subject

    def teach(self):
        print(f"Teacher {self.name} is teaching {self.subject}")


class School:
    def __init__(self):
        self.students = []
        self.teachers = []

    def add_student(self, student):
        self.students.append(student)

    def add_teacher(self, teacher):
        self.teachers.append(teacher)

    def display_school_info(self):
        print("Students:")
        for student in self.students:
            student.display_info()
            student.calculate_average()

        print("\nTeachers:")
        for teacher in self.teachers:
            teacher.display_info()
            teacher.teach()


# Example usage:
# Create instances of Person, Student, Teacher, and School classes
person1 = Person("Jihen", 30)
student1 = Student("Malek", 18, "S123", 85)
teacher1 = Teacher("Sabrine", 40, "T456", "Math")
school = School()

# Add students and teachers to the school
school.add_student(student1)
school.add_teacher(teacher1)

# Display school information
school.display_school_info()
```
